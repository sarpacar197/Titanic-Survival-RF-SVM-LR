```python
tr_data["Title"] = tr_data["Name"].str.extract(r',\s*([^\.]+)\.')
tr_data["Age"] = tr_data["Age"].fillna(tr_data.groupby('Title')["Age"].transform('median'))
```

```python
tr_data['HasCabin'] = tr_data['Cabin'].notnull().astype(int)
tr_data.drop(columns='Cabin', inplace = True)
```

```python
tr_data['Embarked'] = tr_data['Embarked'].fillna(tr_data['Embarked'].mode()[0])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tr_data['Sex'] = tr_data['Sex'].map({'male': 0, 'female': 1})
abs(tr_data.corr(numeric_only = True)["Survived"].sort_values())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
features = ['Sex', 'HasCabin', 'Fare', 'Age', 'Pclass']
X= tr_data[features]
y= tr_data['Survived']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 101)

model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
accuracy_score(y_test, pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model2 = RandomForestClassifier()
```

```python
model2.fit(X_train,y_train)
```

```python
pred2 = model2.predict(X_test)
accuracy_score(y_test, pred2)
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled= scaler.fit_transform(X_test)

model3 = SVC()
model3.fit(X_train_scaled, y_train)
pred3 = model3.predict(X_test_scaled)
```

```python
from sklearn.model_selection import cross_val_score

for name,m,data in [('LogReg', model,X),('RF', model2,X),('SVM', model3, scaler.fit_transform(X))]:
    scores = cross_val_score(m,data,y,cv=5)
    print(name, scores.mean().round(4))
```
